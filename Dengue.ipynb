```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df13 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabela_Oficial_2013.csv', index_col = 0, parse_dates = True, na_values = 0, encoding='latin-1')
df14 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabela_Oficial_2014.csv', parse_dates = True, na_values = 0, encoding='latin-1')
df15 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabela_Oficial_2015.csv', parse_dates = True,  na_values = 0, encoding='latin-1')
df16 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabela_Oficial_2016.csv', parse_dates = True, na_values = 0, encoding='latin-1')
#df = pd.concat([df13, df14, df15, df16])
df13 = df13.fillna(0)
df14 = df14.fillna(0)
df15 = df15.fillna(0)
df16 = df16.fillna(0)
```

```python
def Dengue (arq, ano):
    bairros = np.array(arq['bairro'].unique())
    casos = []
    for bairro in bairros:
        df1 = arq[arq['bairro'] == bairro]
        casos.append(df1['bairro'].count())
    data = {'bairro': bairros, 'casos': casos}
    dengue = pd.DataFrame(data)
    dengue['ano'] = ano
    return dengue
```

```python
df13_aux = df13.sort_values('casos', ascending=True)
df13_sorted = df13_aux.iloc[:55]
plt.figure(figsize=(15,15))
plt.bar(df13_sorted['bairro'], df13_sorted['casos'])
plt.xticks(fontsize=10,rotation=80)
plt.show()
plt.savefig('dg13.jpg',dpi=1000)
```
