```python
def Dengue (arq, ano):
    cod_bairros = np.array(arq['co_bairro_residencia'].unique())
    bairros = np.array(arq['no_bairro_residencia'].unique())
    casos = []
    for cod in cod_bairros:
        df1 = arq[arq['co_bairro_residencia'] == cod]
        casos.append(df1['co_bairro_residencia'].count())
    data = {'no_bairro_residencia':bairros,'co_bairro_residencia': cod_bairros, 'casos': casos}
    dengue = pd.DataFrame(data)
    dengue['co_uf_notificacao'] = ano
    return dengue
```

```python
def dia(a):
    d=datetime.datetime.strptime(a, "%d/%m/%Y").day
    return d
```

```python
def ano(a):
    an=datetime.datetime.strptime(a, "%d/%m/%Y").year
    return an
```

```python
def mes(a):
    m=datetime.datetime.strptime(a, "%d/%m/%Y").month
    return m
```

```python
mes('11/06/2010')
```

```python
def semana(a):
    dt = datetime.date(ano(a), mes(a), dia(a)) 
    wk = dt.isocalendar()[1]
    return wk
```
