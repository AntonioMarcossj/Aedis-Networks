```python
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
```

```python
C13= pd.read_csv('Caso_total_Dengue2013.csv', parse_dates = True, sep = ',', na_values = 0, encoding='latin-1',low_memory=False)
```

```python
casos13=[]
for semana in range(1,53):
    casos13.append(C13[C13['semana_ocorrencia']==semana]['semana_ocorrencia'].count())
```

```python
plt.plot(casos13,color='red',marker='o')
plt.ylabel('Casos Epidêmicos de Dengue em 2013')
plt.xlabel('Semana')
plt.xlim(1,52)
plt.xticks(np.arange(1, 53, 3))
```
