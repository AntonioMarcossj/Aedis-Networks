```python
#Programa que cria os plots do arquivo da dengue
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Tabelas feitas pelo programa Tabalas Oficiais
df13 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabala_Oficial_2013.csv', parse_dates = True, encoding='latin-1')
df14 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabala_Oficial_2014.csv', parse_dates = True, encoding='latin-1')
df15 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabala_Oficial_2015.csv', parse_dates = True, encoding='latin-1')
df16 = pd.read_csv('https://raw.githubusercontent.com/AntonioMarcossj/Aedis-Networks/Tabalas-oficiais/Tabala_Oficial_2016.csv', parse_dates = True, encoding='latin-1')
df13.head()
```

```python
def Grafico(arq, nome, num, boolean, ano):
    #Programa que cria gráficos. Os argumentos são respectivamente:
    # arquivo, tipo de comparação (casos ou casos/pop), número 
    # de bairros a ser visualizado no plot, ordem por número de casos
    # e ano.
    aux = arq.sort_values(nome, ascending = boolean)
    sort = aux.iloc[:num]
    plt.figure(figsize=(15,15))
    plt.bar(sort['Bairro'], sort[nome])
    plt.xticks(fontsize=10,rotation=80)
    if boolean == True:
        plt.title('Os ' + str(num) + ' bairros com menor número de ' + nome +' em ' + str(ano))
    else:
        plt.title('Os ' + str(num) + ' bairros com maior número de ' + nome +' em ' + str(ano))
    plt.show()
```

```python
Grafico(df13, 'Casos', 55, False, 2013)
```

```python
Grafico(df14, 'Casos/Pop', 55, True, 2014)
```
